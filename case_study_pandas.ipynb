#Q8 . Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
df.info()
```

```python
median_value1 = df['Revenue (Millions)'].median()
```

```python
median_value2 = df['Metascore'].median()
```

```python
#replacing null value with median of respective column
df['Revenue (Millions)'].fillna(median_value1,inplace=True)
df['Metascore'].fillna(median_value2,inplace=True)
df.info()
```

#Q9 . Top 5 movies by Rating

```python
top_5_movies = df.sort_values(by='Rating',ascending=False).head(5)
top_5_movies[['Title','Rating']]
```

#Q10 .  Movies with highest/lowest revenue

```python
Highest_revenue = df.sort_values(by='Revenue (Millions)', ascending=False).head(1)
Lowest_revenue = df.sort_values(by='Revenue (Millions)', ascending=True).head(1)

print(" Highest Revenue Movie:\n")
print(Highest_revenue[['Title', 'Revenue (Millions)']].to_string(index=False))

print("\n Lowest Revenue Movie:\n")
print(Lowest_revenue[['Title', 'Revenue (Millions)']].to_string(index=False))
```

#Q16 . Create a new column Profitability:

#Q17 . 17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 â‰¤ Rating < 7.5: Average
c) Rating â‰¥ 7.5: Excellent

```python
df.loc[df['Rating'] < 6, 'Category'] = 'Poor'
print(df[df['Rating'] < 6][['Title', 'Rating', 'Category']])
```

```python
average_movies = df[(df['Rating'] >= 6) & (df['Rating'] < 7.5)][['Title', 'Rating']]
average_movies['Category'] = 'Average'

print(average_movies)
```

```python
excellent_movies = df[df['Rating'] >= 7.5][['Title', 'Rating']]
excellent_movies['Category'] = 'Excellent'

print(excellent_movies)
```

#Q18 .  Create a custom function to score movies based on:
i)
Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(score, axis=1)
df[['Title','High_Score']]
```
