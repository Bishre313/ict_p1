#Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(list1):
  minimum_value=list1[0]
  #set 0 index the minimum value
  maximum_value=list1[0]
  #set 0 indexed value as a maximum value
  for i in list1:
    if(i >= maximum_value):
      maximum_value=i

    if(i <= minimum_value):
      minimum_value=i

  return (minimum_value, maximum_value)


list1=[2,3,5,65,77,7]
range=find_range(list1)
print("maximum and minimum value:",range)
```
